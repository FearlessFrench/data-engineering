```python
import pandas as pd
import numpy as np
```

```python
# If file is not in the same folder as this notebook
# pd.read_csv("C:\\Users\\CSB307\\Desktop\\Download\\Lab_06\\NFL.csv")
nfl_data = pd.read_csv("NFL.csv")
nfl_data.head()
```

**Check missing data**

```python
nfl_data.shape # (rows, cols)
```

```python
# get the number of missing data points per column
missing_values_count = nfl_data.isnull().sum() # คำสั่งในการนับ Missing Value ตามคอลัมน์

# look at the # of missing points in the first ten columns
missing_values_count[0:10] # ตอนนี้เราดูแค่ 10 คอลัมน์ เราจะทำยังไงให้ดู Missing Value ได้ครบทุกคอลัมน์
```

```python
# how many total missing values do we have?
total_cells = np.product(nfl_data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

**Drop missing values**

```python
print("Total Records =", nfl_data.shape[0])
```

```python
# remove all the rows that contain a missing value
nfl_data.dropna() # ลบ rows ที่มี Missing Value
```

```python
# remove all the rows that contain a missing value in specific columns
nfl_data.dropna(subset=['down'])
```

```python
nfl_data.shape # ตอนนี้ inplace = False กล่าวคือ เซตข้อมูลต้นฉบับไม่เปลี่ยนแปลง
```

```python
# remove all columns with at least one missing value
nfl_data.dropna(axis=1)
```

**Impute missing data**

```python
# fill all missing with a value
results = nfl_data.fillna(0) # แทนค่า Missing Value ให้เป็น 0
results.isnull().sum()
```

```python
results.head()
```

```python
# fill missing values in a specific column with a value
results = nfl_data
results['down'] = results['down'].fillna(0) # แทนค่า Missing Value ให้เป็น 0 เฉพาะคอลัมน์ down
results.isnull().sum()
```

```python
results.head()
```

```python
# fill missing values in a specific column with mean
results = nfl_data
results['down'] = results['down'].fillna(results['down'].mean()) # แทนค่า Missing Value ใน down ด้วยค่าเฉลี่ย
results.isnull().sum()
```

ปฏิบัติการข้อ 1) ให้นักศึกษาเติมข้อมูลสูญหายของตัวแปรทั้งหมดที่มีข้อมูลสูญหาย จะใช้วิธีการใดก็ได้ ตามความเหมาะสม พร้อมเขียนอธิบาย

```python
# Answer here!
# ทำการแบ่งการตรวจสอบและจัดการค่าสูญหายทีละ 10 คอลัมน์ เพื่อความสะดวกในการตรวจสอบลักษณะตัวแปรต่างๆ
nfl_data.isnull().sum()[0:10] # Feature ที่ 1-10
nfl_data.isnull().sum()[10:20] # Feature ที่ 11-20
nfl_data.isnull().sum()[20:30] # Feature ที่ 21-30
nfl_data.isnull().sum()[30:40] # Feature ที่ 31-40
nfl_data.isnull().sum()[40:50] # Feature ที่ 41-50
nfl_data.isnull().sum()[50:60] # Feature ที่ 51-60
nfl_data.isnull().sum()[60:70] # Feature ที่ 61-70
nfl_data.isnull().sum()[70:80] # Feature ที่ 71-80
nfl_data.isnull().sum()[80:90] # Feature ที่ 81-90
nfl_data.isnull().sum()[90:] # Feature ที่ 91-102

nfl_data.head() # ลักษณะของชุดข้อมูลในตอนนี้
```

```python
# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 1-10 
nfl_data.isnull().sum()[0:10] # พบ Missing Value ใน Feature: time, TimeSecs, PlayTimeDiff, SideofField

# ตรวจสอบประเภทข้อมูลในคอลัมน์ที่ 1-10
nfl_data.iloc[:, 0:10].dtypes 
# object: time, SideofField
# float64: TimeSecs, PlayTimeDiff

# แทนค่าข้อมูลสูญหาย (Imputing Missing Values)
nfl_data["time"] = nfl_data["time"].ffill() # แทนค่าใน Feature time ด้วยค่าเวลาที่อยู่ก่อนหน้า (Forward Fill) เนื่องจากข้อมูลมีการเรียงลำดับเวลาอยู่ระดับหนึ่ง

nfl_data["TimeSecs"] = nfl_data["TimeSecs"].ffill() # แทนค่าใน Feature TimeSecs ด้วยค่าเวลาที่อยู่ก่อนหน้า (Forward Fill) เนื่องจากข้อมูลมีการเรียงลำดับเวลาอยู่ระดับหนึ่ง

nfl_data["PlayTimeDiff"] = nfl_data["PlayTimeDiff"].fillna(nfl_data["PlayTimeDiff"].mean()//1) # แทนค่าใน Feature TimeSecs ด้วยค่าเฉลี่ย (Mean) ที่ปัดเศษลง เนื่องจากข้อมูลเป็นเลขจำนวนเต็ม แต่เขียนในรูปทศนิยม

nfl_data["SideofField"] = nfl_data["SideofField"].ffill() # แทนค่าใน Feature SideofField ด้วยชื่อสนามที่อยู่ก่อนหน้า (Forward Fill) เนื่องจากข้อมูลมีการเรียงลำดับชื่อสนามอยู่ระดับหนึ่ง

# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 1-10 อีกครั้ง
nfl_data.isnull().sum()[0:10] 
```

```python
# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 11-20 
nfl_data.isnull().sum()[10:20] # พบ Missing Value ใน Feature: yrdln, yrdline100, GoalToGo, FirstDown, posteam, DefensiveTeam

# ตรวจสอบประเภทข้อมูลในคอลัมน์ที่ 11-20
nfl_data.iloc[:, 10:20].dtypes 
# object: posteam, DefensiveTeam
# float64: yrdln, yrdline100, GoalToGo, FirstDown

# แทนค่าข้อมูลสูญหาย (Imputing Missing Values)
nfl_data["yrdln"] = nfl_data["yrdln"].ffill() # แทนค่าใน Feature yrdln ด้วยค่าที่อยู่ก่อนหน้า (Forward Fill) เนื่องจากข้อมูลมีการเรียงลำดับเส้นหลา

nfl_data["yrdline100"] = nfl_data["yrdline100"].ffill() # แทนค่าใน Feature yrdline100 ด้วยค่าที่อยู่ก่อนหน้า (Forward Fill) เนื่องจากข้อมูลมีการเรียงลำดับเส้น 100 หลา

nfl_data["GoalToGo"] = nfl_data["GoalToGo"].fillna(nfl_data["GoalToGo"].mode()[0]) # แทนค่าใน Feature GoalToGo ด้วยค่าฐานนิยม (Mode) เนื่องจากข้อมูลเป็นไปได้แค่ 2 ค่า คือ 0.0 กับ 1.0

nfl_data["FirstDown"] = nfl_data["FirstDown"].fillna(nfl_data["FirstDown"].mode()[0]) # แทนค่าใน Feature FirstDown ด้วยค่าฐานนิยม (Mode) เนื่องจากข้อมูลเป็นไปได้แค่ 2 ค่า คือ 0.0 กับ 1.0

nfl_data["posteam"] = nfl_data["posteam"].ffill() # แทนค่าใน Feature posteam ด้วยค่าที่อยู่ก่อนหน้า (Forward Fill) เนื่องจากข้อมูลมีการเรียงลำดับชื่อทีม

nfl_data["DefensiveTeam"] = nfl_data["DefensiveTeam"].ffill() # แทนค่าใน Feature DefensiveTeam ด้วยค่าที่อยู่ก่อนหน้า (Forward Fill) เนื่องจากข้อมูลมีการเรียงลำดับชื่อทีม

# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 11-20 อีกครั้ง
nfl_data.isnull().sum()[10:20]
```

```python
# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 21-30
nfl_data.isnull().sum()[20:30] # พบ Missing Value ใน Feature: ExPointResult, TwoPointConv, DefTwoPoint, PuntResult

# ตรวจสอบประเภทข้อมูลในคอลัมน์ที่ 21-30
nfl_data.iloc[:, 20:30].dtypes 
# object: ExPointResult, TwoPointConv
# float64: DefTwoPoint, PuntResult

# แทนค่าข้อมูลสูญหาย (Imputing Missing Values) ด้วยคำว่า Missing เนื่องจากข้อมูลใน Feature เหล่านี้ มีค่าสูญหายจำนวนมากเกินไป
nfl_data["ExPointResult"] = nfl_data["GoalToGo"].fillna("Missing") # แทนค่าใน Feature ExPointResult 
nfl_data["TwoPointConv"] = nfl_data["TwoPointConv"].fillna("Missing") # แทนค่าใน Feature TwoPointConv
nfl_data["DefTwoPoint"] = nfl_data["DefTwoPoint"].fillna("Missing") # แทนค่า Feature DefTwoPoint
nfl_data["PuntResult"] = nfl_data["PuntResult"].fillna("Missing") # แทนค่า Feature PuntResult

# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 21-30 อีกครั้ง
nfl_data.isnull().sum()[20:30]
```

```python
# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 31-40
nfl_data.isnull().sum()[30:40] # พบ Missing Value ใน Feature: Passer, Passer_ID, PassOutcome, PassLength, PassLocation

# ตรวจสอบประเภทข้อมูลในคอลัมน์ที่ 31-40
nfl_data.iloc[:, 30:40].dtypes 
# object: Passer, Passer_ID, PassOutcome, PassLength, PassLocation

# แทนค่าข้อมูลสูญหาย (Imputing Missing Values) ด้วยคำว่า Missing เนื่องจากข้อมูลใน Feature เหล่านี้ ไม่ได้ระบุชื่อหรือรหัสของผู้ส่งบอล จึงถือว่าไม่ได้ทำการเก็บข้อมูลไว้
nfl_data["Passer"] = nfl_data["Passer"].fillna("Missing") # แทนค่าใน Feature Passer
nfl_data["Passer_ID"] = nfl_data["Passer_ID"].fillna("Missing") # แทนค่าใน Feature Passer_ID
nfl_data["PassOutcome"] = nfl_data["PassOutcome"].fillna("Missing") # แทนค่า Feature PassOutcome
nfl_data["PassLength"] = nfl_data["PassLength"].fillna("Missing") # แทนค่า Feature PassLength
nfl_data["PassLocation"] = nfl_data["PassLocation"].fillna("Missing") # แทนค่า Feature PassLocation

# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 31-40 อีกครั้ง
nfl_data.isnull().sum()[30:40]
```

```python
# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 41-50
nfl_data.isnull().sum()[40:50] # พบ Missing Value ใน Feature: Interceptor, Rusher, Rusher_ID, RunLocation, RunGap, Receiver, Receiver_ID, ReturnResult

# ตรวจสอบประเภทข้อมูลในคอลัมน์ที่ 41-50
nfl_data.iloc[:, 40:50].dtypes 
# object: Interceptor, Rusher, Rusher_ID, RunLocation, RunGap, Receiver, Receiver_ID, ReturnResult

# แทนค่าข้อมูลสูญหาย (Imputing Missing Values) ด้วยคำว่า Missing เนื่องจากข้อมูลใน Feature เหล่านี้ มีค่าสูญหายจำนวนมากเกินไป
nfl_data["Interceptor"] = nfl_data["Interceptor"].fillna("Missing") # แทนค่าใน Feature Interceptor
nfl_data["Rusher"] = nfl_data["Rusher"].fillna("Missing") # แทนค่าใน Feature Rusher
nfl_data["Rusher_ID"] = nfl_data["Rusher_ID"].fillna("Missing") # แทนค่า Feature Rusher_ID
nfl_data["RunLocation"] = nfl_data["RunLocation"].fillna("Missing") # แทนค่า Feature RunLocation
nfl_data["RunGap"] = nfl_data["RunGap"].fillna("Missing") # แทนค่า Feature RunGap
nfl_data["Receiver"] = nfl_data["Receiver"].fillna("Missing") # แทนค่า Feature Receiver
nfl_data["Receiver_ID"] = nfl_data["Receiver_ID"].fillna("Missing") # แทนค่า Feature Receiver_ID
nfl_data["ReturnResult"] = nfl_data["ReturnResult"].fillna("Missing") # แทนค่า Feature ReturnResult

# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 41-50 อีกครั้ง
nfl_data.isnull().sum()[40:50]
```

```python
# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 51-60
nfl_data.isnull().sum()[50:60] 
# พบ Missing Value ใน Feature: Returner, BlockingPlayer, Tackler1, Tackler2, FieldGoalResult, FieldGoalDistance, RecFumbTeam, RecFumbPlayer

# ตรวจสอบประเภทข้อมูลในคอลัมน์ที่ 51-60
nfl_data.iloc[:, 50:60].dtypes 
# object: Returner, BlockingPlayer, Tackler1, Tackler2, FieldGoalResult, RecFumbTeam, RecFumbPlayer
# float64: FieldGoalDistance

# แทนค่าข้อมูลสูญหาย (Imputing Missing Values) ด้วยคำว่า Missing เนื่องจากข้อมูลใน Feature เหล่านี้ มีค่าสูญหายจำนวนมากเกินไป
nfl_data["Returner"] = nfl_data["Returner"].fillna("Missing") # แทนค่าใน Feature Returner
nfl_data["BlockingPlayer"] = nfl_data["BlockingPlayer"].fillna("Missing") # แทนค่าใน Feature BlockingPlayer
nfl_data["Tackler1"] = nfl_data["Tackler1"].fillna("Missing") # แทนค่า Feature Tackler1
nfl_data["Tackler2"] = nfl_data["Tackler2"].fillna("Missing") # แทนค่า Feature Tackler2
nfl_data["FieldGoalResult"] = nfl_data["FieldGoalResult"].fillna("Missing") # แทนค่า Feature FieldGoalResult
nfl_data["FieldGoalDistance"] = nfl_data["FieldGoalDistance"].fillna("Missing") # แทนค่า Feature FieldGoalDistance
nfl_data["RecFumbTeam"] = nfl_data["RecFumbTeam"].fillna("Missing") # แทนค่า Feature RecFumbTeam
nfl_data["RecFumbPlayer"] = nfl_data["RecFumbPlayer"].fillna("Missing") # แทนค่า Feature RecFumbPlayer

# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 51-60 อีกครั้ง
nfl_data.isnull().sum()[50:60]
```

```python
# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 91-102
nfl_data.isnull().sum()[90:]
# พบ Missing Value ใน Feature: EPA, airEPA, yacEPA, Home_WP_pre, Away_WP_pre, Home_WP_post, Away_WP_post, Win_Prob, WPA, airWPA, yacWPA

# ตรวจสอบประเภทข้อมูลในคอลัมน์ที่ 91-102
nfl_data.iloc[:, 90:].dtypes 
# float64: EPA, airEPA, yacEPA, Home_WP_pre, Away_WP_pre, Home_WP_post, Away_WP_post, Win_Prob, WPA, airWPA, yacWPA

# แทนค่าข้อมูลสูญหาย (Imputing Missing Values)
# แทนค่าของ Feature ที่เป็นค่าความน่าจะเป็นด้วยค่าเฉลี่ย (Mean)
nfl_data["EPA"] = nfl_data["EPA"].fillna(nfl_data["EPA"].mean()) # แทนค่าใน Feature EPA
nfl_data["Home_WP_pre"] = nfl_data["Home_WP_pre"].fillna(nfl_data["Home_WP_pre"].mean()) # แทนค่าใน Feature Home_WP_pre
nfl_data["Away_WP_pre"] = nfl_data["Away_WP_pre"].fillna(nfl_data["Away_WP_pre"].mean()) # แทนค่าใน Feature Away_WP_pre
nfl_data["Home_WP_post"] = nfl_data["Home_WP_post"].fillna(nfl_data["Home_WP_post"].mean()) # แทนค่าใน Feature Home_WP_post
nfl_data["Away_WP_post"] = nfl_data["Away_WP_post"].fillna(nfl_data["Away_WP_post"].mean()) # แทนค่าใน Feature Away_WP_post
nfl_data["Win_Prob"] = nfl_data["Win_Prob"].fillna(nfl_data["Win_Prob"].mean()) # แทนค่าใน Feature Win_Prob
nfl_data["WPA"] = nfl_data["WPA"].fillna(nfl_data["WPA"].mean()) # แทนค่าใน Feature Win_Prob

# แทนค่าด้วยคำว่า Missing เนื่องจากมีค่าสูญหายจำนวนมากเกินไป
nfl_data["airEPA"] = nfl_data["airEPA"].fillna("Missing") # แทนค่า Feature airEPA 
nfl_data["yacEPA"] = nfl_data["yacEPA"].fillna("Missing") # แทนค่าใน Feature yacEPA
nfl_data["airWPA"] = nfl_data["airWPA"].fillna("Missing") # แทนค่า Feature airWPA
nfl_data["yacWPA"] = nfl_data["yacWPA"].fillna("Missing") # แทนค่าใน Feature yacWPA

# ตรวจสอบข้อมูลสูญหายในคอลัมน์ที่ 91-102 อีกครั้ง
nfl_data.isnull().sum()[90:]
```
