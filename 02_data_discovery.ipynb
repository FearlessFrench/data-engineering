### Data Quality Check

#### Completeness
The values of some components of a single datum are still invalid

```python
data.isnull().sum()
```

```python
data[data["country_name"].isnull() == True] # อาจแทนด้วย Bangladesh (ดูจาก country_code)
```

#### Uniqueness
มีข้อมูลที่ซ้ำกันมากกว่า 1 แถวอยู่ 100 ค่า

```python
data.duplicated().sum()
```

```python
data[data.duplicated()]
```

```python
data[data["country_name"] == "Algeria"][data["indicator_code"] == "RL.STD.ERR"] # ตัวอย่างของการซ้ำข้อมูล
```

#### Consistency
มีข้อมูลที่ไม่มีความคงเส้นคงวาอยู่ Bahamas, The มีรูปแบบความแตกต่างจากชื่อประเทศอื่นๆ ในชุดข้อมูล ควรแก้ให้เป็นชื่อทางการซึ่งก็คือ The Bahamas

```python
data["country_name"].unique()
```

```python
data[data["country_name"] == "Bahamas, The"].head()
```

#### Validity
ข้อมูลที่ใช้ในการคำนวณหรือประมาณค่าอยู่ในช่วงและค่าที่ถูกต้อง บางข้อมูลอาจจะต้องแก้ค่า NaN ให้เป็น 0 ก่อน

```python
data.indicator_code.unique()
```

```python
data[data.indicator_code == "CC.EST"].head() # การควบคุมอัตราการทุจริต ถ้าค่าเป็น + คือควบคุมได้ดีมาก

data[data.indicator_code == "GE.NO.SRC"].head() # จำนวนแหล่งข้อมูลที่ใช้ในการวิเคราะห์ ถ้าไม่มีแสดงค่า NaN

data[data.indicator_code == "PV.PER.RNK"].head() # ความมั่นคงของภาครัฐ ค่า Percentile Rank ตั้งแต่ 0-100

data[data.indicator_code == "RQ.PER.RNK.LOWER"].head() # คุณภาพกฎหมาย พบว่ามีค่าตกที่ 0.000000

data[data.indicator_code == "RL.PER.RNK.UPPER"].head() # กฎหมาย พบว่ามีค่าตกที่ 100.000000

data[data.indicator_code == "VA.STD.ERR"].head() # สิทะิเสรีภาพของประชาชน แสดงค่า Standard Error
```

#### Accuracy
เท่าที่ดูจาก Range ของแต่ละค่าที่เป็นไปได้ ข้อมูลนี้มีความแม่นยำสูงอยู่

อาจจะยังต้องแก้บางค่าที่เป็น NaN ให้เป็น 0 อยู่เช่น จำนวนแหล่งข้อมูลที่ใช้ในการวิเคราะห์ข้อมูล

```python
data.groupby('indicator_code')['1996'].agg(['min', 'max']).head(6)
```

```python
data.groupby('indicator_code')['2012'].agg(['min', 'max']).head(6)
```
